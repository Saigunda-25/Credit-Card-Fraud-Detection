```python
#Importing the Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
```

```python
#Reading the Data 

credit_data=pd.read_csv("C:/Users/balas/Downloads/Credit Card Fraud Detection/creditcard.csv")
```

```python
credit_data.head()
```

```python
print(credit_data.shape)
print(credit_data['Class'].value_counts())

# Check for imbalance
sns.countplot(x='Class', data=credit_data)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()
```

```python
#Standardising the Data
scaler = StandardScaler()
credit_data['scaled_amount'] = scaler.fit_transform(credit_data[['Amount']])
credit_data = credit_data.drop(['Time', 'Amount'], axis=1)
```

```python
#Splitting the Dataset into Training and Testing Part
X = credit_data.drop('Class', axis=1)
y = credit_data['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
#Building the Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
#Prediction on the Testing Data
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
#ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.figure(figsize=(8,5))
plt.plot(fpr, tpr, label='Logistic Regression')
plt.plot([0,1], [0,1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid()
plt.show()
```
